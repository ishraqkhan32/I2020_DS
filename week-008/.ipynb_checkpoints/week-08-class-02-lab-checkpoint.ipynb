Write the code that accesses the h3 tag in the following html.

```python
from bs4 import BeautifulSoup as bs

html = \
"""
<html>
  <head>
    <title>our webpage</title>
  </head>
  <body>
    <header>
      <h1>hello world</h1>
      <div>
        <h3>tag line</h3>
      </div>
    </header>
    <p>paragraph</p>
    <ul>
      <li><a href='#' id='link1'>link 1</a></li>
      <li><a href='#' id='link2'>link 2</a></li>
      <li><a href='#' id='link3'>link 3</a></li>
    </ul>
  </body>
</html>
"""

# write your code below:
soup = bs(html)

h3 = soup.find('h3')
h3.contents
```

# Exercise 2

Create a function that will take a list of web pages and return a list of all the anchor tags on those pages.

```python
import requests
```

```python
links = ['https://www.google.com/', 'https://www.nytimes.com/', 'https://www.youtube.com/']

def get_anchors(links):
    anchor_list = []
    
    for url in links:
        # convert each url into soup
        r = requests.get(url)
        soup = bs(r.text)
        
        # get list of all anchors
        a_tags = soup.find_all('a')
    
        # filter a tags
        for anchor in a_tags:
            try:
                if (link := anchor['href']).startswith('http'):
                    anchor_list.append(link)
                else:
                    anchor_list.append(url + link)
            except KeyError:
                continue
                
    return anchor_list

get_anchors(links)
```
